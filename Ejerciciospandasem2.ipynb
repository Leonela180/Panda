```python
#Ejercicio 1:
#rango de años
inicio = int(input("Ingrese el año de inicio: "))
fin = int(input("Ingrese su año final: "))

# ventas para cada año
ventas = {}
for anio in range(inicio, fin + 1):
    monto = float(input(f"Ingrese las ventas del año {anio}: "))
    ventas[anio] = monto

# serie original
serie_ventas = pd.Series(ventas)
print("\nVentas originales:")
print(serie_ventas)

# descuento del 10%
ventas_con_descuento = serie_ventas * 0.9
print("\nVentas con 10% de descuento:")
print(ventas_con_descuento)
```

```python
#Ejercicio 2:
def resumen_notas(notas_dict):
    notas = pd.Series(notas_dict)

    resumen = pd.Series({
        'Nota mínima': notas.min(),
        'Nota máxima': notas.max(),
        'Media': notas.mean(),
        'Desviación típica': notas.std()
    })

    return resumen
notas_alumnos = {
    'Ana': 15,
    'Luis': 18,
    'Pedro': 12,
    'Lucía': 17,
    'Carlos': 14
}

resultado = resumen_notas(notas_alumnos)
print(resultado)
```

```python
#Ejercicio 3:

def alumnos_aprobados(notas_dict):
    notas = pd.Series(notas_dict)

    aprobados = notas[notas >= 11]

    # Ordenar de mayor a menor
    aprobados_ordenados = aprobados.sort_values(ascending=False)

    return aprobados_ordenados

# Ejemplo de uso
notas_alumnos = {
    'Ana': 15,
    'Luis': 18,
    'Pedro': 12,
    'Lucía': 17,
    'Carlos': 14,
    'José': 10
}

resultado = alumnos_aprobados(notas_alumnos)
print(resultado)
```

```python
#Ejercicio 5:
def calcular_balance(df, meses):
    # Filtrar los meses indicados
    df_filtrado = df[df['Mes'].isin(meses)]

    # Calcular balance por mes y sumarlo
    balance_total = (df_filtrado['Ventas'] - df_filtrado['Gastos']).sum()

    return balance_total

# Datos
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
    'Ventas': [30500, 35600, 28300, 33900],
    'Gastos': [22000, 23400, 18100, 20700]
}
df = pd.DataFrame(datos, columns=['Mes', 'Ventas', 'Gastos'])

# Ejemplo de uso
meses_consulta = ['Enero', 'Marzo']
balance = calcular_balance(df, meses_consulta)
print(f"Balance total en {meses_consulta}: {balance}")
```

```python
#Ejercicio 6:
def resumen_columna_cotizacion(ruta_archivo, columna):
    try:
        # Leer el archivo CSV
        df = pd.read_csv(ruta_archivo)

        # Verificar si la columna existe
        if columna not in df.columns:
            raise ValueError(f"La columna '{columna}' no existe en el archivo.")

        # Calcular mínimo, máximo y media
        resumen = {
            'Mínimo': df[columna].min(),
            'Máximo': df[columna].max(),
            'Media': df[columna].mean()
        }

        # Convertir a DataFrame
        resumen_df = pd.DataFrame(resumen, index=[columna])

        return resumen_df

    except FileNotFoundError:
        print(f"El archivo '{ruta_archivo}' no se encontró.")
    except Exception as e:
        print(f"Ocurrió un error: {e}")
    resultado = resumen_columna_cotizacion("cotizacion.csv", "Final")
print(resultado)
```
