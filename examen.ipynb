```python
# 1. Crea un arreglo NumPy de dimensión 1 con los números del 1 al 20.
arreglo_1_a_20 = np.arange(1, 21)
print("Arreglo del 1 al 20:")
display(arreglo_1_a_20)
```

```python
# 3. Genera una matriz aleatoria de tamaño 5x5 con números entre 0 y 1.
matriz_aleatoria = np.random.rand(5, 5)
print("\nMatriz aleatoria 5x5:")
display(matriz_aleatoria)
```

```python
# 4. Calcula la media, la desviación estándar y el valor máximo de esa matriz.
media_matriz = np.mean(matriz_aleatoria)
desviacion_estandar_matriz = np.std(matriz_aleatoria)
maximo_matriz = np.max(matriz_aleatoria)

print(f"\nMedia de la matriz: {media_matriz}")
print(f"Desviación estándar de la matriz: {desviacion_estandar_matriz}")
print(f"Valor máximo de la matriz: {maximo_matriz}")
```

```python
# 1. Crea un DataFrame con los siguientes datos de estudiantes:
data = {'Nombre': ['Ana', 'Luis', 'María', 'Pedro', 'Sofía'],
        'Edad': [20, 21, 19, 22, 20],
        'Nota_Matemáticas': [18, 15, 16, 12, 19],
        'Nota_IA': [17, 14, 18, 13, 20]}
df_estudiantes = pd.DataFrame(data)
print("DataFrame de estudiantes:")
display(df_estudiantes)
```

```python
 #Ejercicio 1
 def analyze_list(numbers):
    if not numbers:
        return None, None, None  # Manejar el caso de lista vacía

    maximum = max(numbers)
    minimum = min(numbers)
    average = sum(numbers) / len(numbers)

    return maximum, minimum, average

# Lista de ejemplo
sample_numbers = [10, 20, 30, 40, 50]

# Llamar a la función e imprimir los resultados
max_val, min_val, avg_val = analyze_list(sample_numbers)

print(f"Lista de ejemplo: {sample_numbers}")
print(f"Máximo: {max_val}")
print(f"Mínimo: {min_val}")
print(f"Promedio: {avg_val}")
```

```python
#Ejercicio 2
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

# Crear el arreglo de entrada
x = np.array([-2, -1, 0, 1, 2])

# Aplicar la función sigmoide al arreglo
sigmoid_x = sigmoid(x)

# Imprimir los arreglos original y resultante
print("Arreglo original:", x)
print("Arreglo activado por sigmoide:", sigmoid_x)
```

```python
# Ejercicio 1: Resolver ecuaciones lineales
def solve_linear(a, b):
    # Resuelve la ecuación lineal ax + b = 0
    if a == 0:
        # Manejar el caso donde a es cero
        return "Esta no es una ecuación lineal o no tiene solución única."
    else:
        # Calcular el valor de x
        x = -b / a
        return x

# Casos de prueba
print(f"Resolviendo 2x + 4 = 0: x = {solve_linear(2, 4)}")
print(f"Resolviendo 0x + 5 = 0: {solve_linear(0, 5)}")
print(f"Resolviendo 3x - 9 = 0: x = {solve_linear(3, -9)}")
```

```python
import cmath # Importar cmath para soporte de números complejos

# Ejercicio 2: Resolver ecuaciones de segundo grado
def solve_quadratic(a, b, c):
    # Calcular el discriminante (delta)
    delta = (b**2) - 4*(a*c)

    # Encontrar las dos soluciones
    x1 = (-b - cmath.sqrt(delta)) / (2*a)
    x2 = (-b + cmath.sqrt(delta)) / (2*a)

    return x1, x2

# Casos de prueba
print("Resolviendo x² - 5x + 6 = 0:")
sol1, sol2 = solve_quadratic(1, -5, 6)
print(f"Soluciones: x1 = {sol1}, x2 = {sol2}")

print("\nResolviendo x² + 2x + 1 = 0:")
sol1, sol2 = solve_quadratic(1, 2, 1)
print(f"Soluciones: x1 = {sol1}, x2 = {sol2}")

print("\nResolviendo x² + x + 1 = 0:")
sol1, sol2 = solve_quadratic(1, 1, 1)
print(f"Soluciones: x1 = {sol1}, x2 = {sol2}")
```
