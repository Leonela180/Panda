SERIES

```python
#Crear una serie
serie=pd.Series([2,6,-1,8,9,2.3], index=['a','b','c','d','e','f'])
serie
```

```python
#borrar serie o filas
serie2=serie.drop(['a','e'])
serie2
```

```python
#seleccionar filas o series
serie[['b','d']]
```

```python
#asignacion
serie[['b']]=100
serie
```

```python
#operaiones
serie4 = (serie+serie2)
serie4
```

```python
#MULTIPLICACION
serie4 = serie.mul(serie2)
serie4
```

```python
#division
serie4 = serie.div(serie2)
serie4
```

```python
#resta
serie4 = serie.sub(serie2)
serie4
```
